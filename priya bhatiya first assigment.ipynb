```python
### 1.
#maincode to check duplicate
#time complaxity O(n^2) reasone 2 loop
#space complaxity O(1)  reasone  no extra space taken
#first define function
def duplicater(arr):
    #conditin of duplicate
    for i in range(len(arr)):
        for j in range(i+1,len(arr)):
            if arr[i] == arr[j]:
                return True
    return False  
            
#drive code
#giving array
arr = [1, 2, 4, 2, 5, 9]
#consider result
result = duplicater(arr)
#now printing result
print(result)
```

```python
##2.2. Given an array and an integer k, rotate the array to the right by k steps.
#def the roteter function
def roteter(arr):
    #K IS ALREADY GIVEN
    k =3
    n = len(arr)
    # because k%n mean 3
    k = k % n

    # ROTATE K ELEMENT FIRST
    arr[:k] = arr[:k][::-1]

    # ROTAT LEFT ARR
    arr[k:] = arr[k:][::-1]

    # ROTATE WHOLE ARR NOW
    arr = arr[::-1]
# now return the  WHOLE ROTATE arr now
    return arr

#drive code
arr = [1, 2, 3, 4, 5, 6, 7]
#Output = [5, 6, 7, 1, 2, 3, 4]
#concider result 
result=roteter(arr)
print(result)
```

```python
#3. Reverse the given array in-place, means without using any extra data structure. 
# def the  Reverse function
#time complaxity O(n)
#space complaxity O(n)
def  Reverse(arr):
    for i in range(len(arr)//2):
        # write logic overhere
        arr[i],arr[-i-1] =arr[-i-1],arr[i]
        # rteuen arr now 
    return arr
        
#drive code
arr = [2, 4, 5, 7, 9, 12] 
result = Reverse(arr)
#Output = [12, 9, 7, 5, 4, 2]
print(result)
```

```python
#4. Given an array of integers, find the maximum element in an array
#first def the functio max_a to find max value
#time complaxity O(n)
#space complaxity O(1)
def max_a(arr,n):
    # maximum element
    max=arr[0]
    #checking every element is big or not comparing other
    # current max element
    for i in range(1,n):
        if arr[i] > max:
            max = arr[i]
            # rteuen arr now 
    return max
#drive code       
arr = [10, 5, 20, 8, 15]
n = len(arr)
result = max_a(arr,n)
print(result)   
#Output = 20 
```

```python
# Python3 program to
# remove duplicates
# Function to remove
# duplicate elements
 
# This function returns new size of modified array
def removeDuplicates(arr, n):
 
    # Return, if array is empty or contains
    # a single element
    if n == 0 or n == 1:
        return n
 
    temp = list(range(n))
 
    # Start traversing elements
    j = 0
    for i in range(0, n-1):
 
        # If current element is not equal to next
        # then store that current element
        if arr[i] != arr[i+1]:
            temp[j] = arr[i]
            j += 1
 
    # Store the last element as whether it is unique
    # or repeated, it isn't stored previously
    temp[j] = arr[n-1]
    j += 1
 
    # Modify original array
    for i in range(0, j):
        arr[i] = temp[i]
 
    return j
 
 # Driver code
if __name__ == '__main__':
    arr = [1, 2, 2, 3, 4, 4, 4, 5, 5]
    n = len(arr)
 
    # removeDuplicates() returns new size of array.
    n = removeDuplicates(arr, n)
 
    # Print updated array
    for i in range(n):
        print("%d" % (arr[i]), end=" ")
 
 
# This code is contributed by aadityaburukwale.
```
